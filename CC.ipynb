```

data =pd.read_csv("glaston29_tf.csv")
data = data[data['hashtags'].notnull()]
data['tweet'] = data['tweet'].str.lower() 
data['hashtags'] = data['hashtags'].str.lower() 
data = data[data.apply(lambda row: len(row.hashtags.split()) > 1, axis=1)]



graph_G = {}
for index, row in data.iterrows():
  edges = list(itertools.combinations(row['hashtags'].split(), 2))
  for edge in edges:
    edge = tuple(sorted(edge))
    if edge not in graph_G.keys():
      graph_G[edge] = 1
    else:
      graph_G[edge] += 1

```

PageRanks

```
# we will take the top 10% pageranks 
import math
# PageRank for largestcc_G
from networkx.algorithms.link_analysis.pagerank_alg import pagerank

def top_10 (lst):
  list_pg=[]
  for i in range((math.ceil(len(lst)*0.1))) :
   list_pg.append(lst[i])
  return (list_pg)
  
def averagepg( l):
 sum=0
 avg=0
 for i in l:
  sum=sum+i
 avg = sum/len(l)
 return avg
```

```

pager=pagerank(largestcc_G, 0.85)

val= [v for k , v in pager.items()]

# sort descending orde the values of pagerank 
sorted_d= sorted((value, key) for (key,value) in pager.items())
sorted_d.sort(reverse=True)
print(sorted_d)

pg=top_10(sorted_d)
pgG=[i[0] for i in pg]  


avgG= averagepg(pgG)
print(avgG )
```

```
pager_erdos=pagerank(largestcc_rnd, 0.85)

val= [v for k , v in pager_erdos.items()]

# sort descending orde the values of pagerank 
sorted_dE= sorted((value, key) for (key,value) in pager_erdos.items())
sorted_dE.sort(reverse=True)
print(sorted_dE)

pgr=top_10(sorted_dE)
pgR=[i[0] for i in pgr]  


avgR= averagepg(pgR)
print(avgR )
```
